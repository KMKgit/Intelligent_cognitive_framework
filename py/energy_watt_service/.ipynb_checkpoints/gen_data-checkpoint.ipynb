```python
import pandas as pd
import numpy as np
import random
from IPython.core.debugger import Tracer 
```

```python
appl_list = ['fan', 'massager', 'hairdryer']
cols = appl_list[:]
cols.append('watt')
output_dim = len(appl_list)
shuffle_num = 50
case_num = pow(2,output_dim)
case_num_list = case_num*[5000]
watt_list = [0, 12, 30, 59, 38, 91, 66, 98]
noise_list = [0, 1, 2, 1, 1, 2, 3, 2]

dir_save = 'data/fan_mix_hair'
path_save = '%s/test_%d_shuffled.csv'%(dir_save, shuffle_num)

if len(watt_list)!= case_num:
    raise IOError("watt_list length error")
if len(case_num_list)!= case_num:
    raise IOError("case_num_list length error")
```

```python
def shuffle_df(df, shuffle_num):
    df = df.reset_index(drop=True)
    total_num = len(df)
    shuffle_idx = df.index.values
    if total_num%shuffle_num != 0:
            pad_num = total_num%shuffle_num
            pad_np = np.zeros(shuffle_num-pad_num)-1
            shuffle_idx = np.append(shuffle_idx, pad_np)
    shuffle_idx = shuffle_idx.reshape((-1, shuffle_num))
    shuffle_idx = np.random.permutation(shuffle_idx)
    shuffle_idx = shuffle_idx.reshape((len(shuffle_idx)*shuffle_num,))
    
    df = df.iloc[shuffle_idx]
    df = df.reset_index(drop=True)
    return df
```

```python
for i in range(case_num):
    tmp_np = np.zeros((case_num_list[i], output_dim+1))   
    tmp_df = pd.DataFrame(tmp_np, columns=cols) 
    bin_str = bin(i)[2:].zfill(output_dim)
    query_list = []
    for j in range(len(bin_str)): 
        tmp_df[appl_list[j]] = int(bin_str[j])
    tmp_df['watt'] = watt_list[i]
    noise_np = np.random.randint(noise_list[i]+1, size=(case_num_list[i]))
    noise_df = pd.DataFrame(noise_np, columns=['watt'])
    tmp_df['watt'] = tmp_df['watt']+noise_df['watt']
    
    if i==0:
        result_df = tmp_df
    else:
        result_df = pd.concat([result_df, tmp_df], axis=0)
```

```python
result_df
```

```python
result_df = shuffle_df(result_df, shuffle_num)
```

```python
result_df
```

```python
result_df.to_csv(path_save, sep=',', index=False)
```
